# Without arguments & with return values

```python
def multiplication_of_two_numbers():
    i=int(input())
    j=int(input())
    k=i*j
    return k
a=multiplication_of_two_numbers()
print(a)
```

# With arguments & without return values

```python
def addition_of_two_numbers(i,j):
    k=i+j
    print(k)
```

```python
i = int(input())
j = int(input())
addition_of_two_numbers(i,j)
```

```python
def multiplication_of_two_numbers(i,j):
    k=i*j
    print(k)
    
```

```python
i = int(input())
j = int(input())
multiplication_of_two_numbers(i,j)
```

# With arguments & return values

```python
def mul_of_two_numbers(a,b):
    c = a*b
    return c
```

```python
a = int(input())
b = int(input())
c= mul_of_two_numbers(a,b)
print(c)
```
